```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

features_cols = ['purpose', 'int_rate', 'installment', 'annual_inc', 'verified', 'home_ownership', 'grdCtoA']
features_data_train1 = train1[features_cols]
label = ['not_paid']
label_data_train1 = train1[label]

# Initialize our classifier
logress = LogisticRegression()
# Train our classifier
model = logress.fit(features_data_train1, label_data_train1)

test1 = pd.read_csv("test1.csv")
features_data_test1 = test1[features_cols]
label_data_test1 = test1[label]

# Make predictions
preds = logress.predict(features_data_test1)
print(preds)
# Evaluate accuracy

print(accuracy_score(label_data_test1, preds))
print("\n")
print(confusion_matrix(label_data_test1,preds))
print("\n")
print(precision_score(label_data_test1,preds))
print("\n")
print(recall_score(label_data_test1,preds))
```
